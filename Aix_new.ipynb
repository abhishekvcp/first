```python
import pandas as pd



def abhi_string_to_numeric(arg1):
    return int(arg1)
          
def abhi_obt_to_string_to_float(arg1):
    new_val = arg1.strip().replace("%","")
    return float(new_val)

def abhi_obt_to_string(arg1):
    new_val = arg1.strip()
    return str(new_val)

custom_fields = ["name","date","time","type","cpupct","availmem"]

xx = pd.read_csv("AIX_Monthly_Application memory_Aug_2019.csv", header=16,skip_blank_lines=True,low_memory=False)

xx.columns = custom_fields

xx.head()

# filter out records only which type "LPAR". this will automatically discard most unwanted records

xx = xx[xx['type'] == 'LPAR']

# anderton lpar had - against memory value. hence it was not getting converted to float type

xx = xx[xx['availmem'] != "-"]

xx.to_csv("test.csv")

xx.info()

xx['cpupct'] = xx['cpupct'].apply(abhi_obt_to_string_to_float)
xx['availmem'] = xx['availmem'].apply(abhi_obt_to_string_to_float)

# create a new field  to display values for Memory% used 

xx['memrealpct'] = 100 - xx['availmem']

xx.info()

xx.head()



```

```python
cpu = xx.groupby(['name'])['cpupct'].quantile([.95])
cpu = pd.DataFrame(cpu)
cpu.info()
cpu = cpu.sort_values(by='cpupct',ascending=False)

print(cpu.head(10))

rmem = xx.groupby(['name'])['memrealpct'].quantile([.8])
rmem = pd.DataFrame(rmem)
rmem = rmem.sort_values(by='memrealpct',ascending=False)

rmem.to_csv("mem_actions.csv")
cpu.to_csv("cpu_actions.csv")
```
